```python
#@title Multiple attack roller for D&D 5E

#@markdown ### Attack Rolls:
attacks = 1 #@param {type:"integer"}
advantage = 'none' #@param ["none", "advantage", "disadvantage"]
attack_bonus = 0 #@param {type:"integer"}
armor_class = 10 #@param {type:"integer"}

#@markdown ### Attack Damage:
resistance = 'none' #@param ["none", "resistance", "vulnerable"]
damage_bonus = 0 #@param {type:"integer"}
d4 = 0 #@param {type:"integer"}
d6 = 0 #@param {type:"integer"}
d8 = 0 #@param {type:"integer"}
d10 = 0 #@param {type:"integer"}
d12 = 0 #@param {type:"integer"}
d20 = 0 #@param {type:"integer"}

#@markdown ### Concentration Check:
check_concentration = False #@param {type:"boolean"}
concentration_advantage = 'none' #@param ["none", "advantage", "disadvantage"]
constitution_save = 0 #@param {type:"integer"}

def roll_damage(num_dice, dice_size):
  damage = 0
  for i in range(num_dice):
    damage += random.randint(1, dice_size)
  return damage

total_damage = misses = hits = crits = 0
hits_record = []

import random

for i in range(attacks):
  attack_roll = random.randint(1, 20)
  if advantage == "advantage":
    attack_roll = max(attack_roll, random.randint(1, 20))
  elif advantage == "disadvantage":
    attack_roll = min(attack_roll, random.randint(1, 20))
  if attack_roll == 1 or attack_roll + attack_bonus < armor_class:
    misses +=1
  else:
    damage = damage_bonus
    damage += roll_damage(d4, 4)
    damage += roll_damage(d6, 6)
    damage += roll_damage(d8, 8)
    damage += roll_damage(d10, 10)
    damage += roll_damage(d12, 12)
    damage += roll_damage(d20, 20)
    if attack_roll == 20:
      damage += roll_damage(d4, 4)
      damage += roll_damage(d6, 6)
      damage += roll_damage(d8, 8)
      damage += roll_damage(d10, 10)
      damage += roll_damage(d12, 12)
      damage += roll_damage(d20, 20)
      crits += 1
    hits += 1
    if resistance == "resistance":
      damage = damage//2
    elif resistance == "vulnerable": 
      damage *= 2
    total_damage += damage
    hits_record.append(damage)

print(f"{total_damage} damage ({hits} attacks hit, {crits} of which were critical hits. {misses} attacks missed.)")
if hits > 0:
  damage_report = "Individual attack damages:"
  for i in hits_record:
    damage_report += " "
    damage_report += str(i)
  damage_report += "."
  print(damage_report)

if check_concentration and hits > 0:
  concentration_broken = False
  concentration_rolls = []
  for i in hits_record:
    save_dc = max(10, i//2)
    save_roll = random.randint(1,20)
    if concentration_advantage == "advantage":
      save_roll = max(save_roll, random.randint(1, 20))
    elif concentration_advantage == "disadvantage":
      save_roll = min(save_roll, random.randint(1, 20))
    save_roll += constitution_save
    concentration_rolls.append(save_roll)
    if save_roll < save_dc:
      concentration_broken = True
  if concentration_broken:
    print("Concentration was broken.")
  else:
    print("Concentration was maintained.")
  concentration_report = "Individual concentration save rolls:"
  for i in range(len(concentration_rolls)):
    concentration_report += f" {str(concentration_rolls[i])} (DC {str(max(10, hits_record[i]//2))})"
  concentration_report += "."
  print(concentration_report)
```
